```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# 1. Prepare a list of offices sorted by country, state, city.

```python
run ("SELECT country,state,city FROM offices ORDER BY country,state,city")
```

# 2. How many employees are there in the company?

```python
run("SELECT COUNT(*)AS NumberofEmployees FROM employees")
```

# 3. What is the total of payments received?

```python
#3.1
run("SELECT COUNT(*)AS NumberofPayments FROM payments")
```

```python
#3.2
run("SELECT SUM(amount) AS TotalPayments FROM payments")
```

# 4. List the product lines that contain 'Cars'.

```python
run("SELECT productLine,productCode,productName FROM products WHERE productLine LIKE '%Cars%'")
```

# 5. Report total payments for October 28, 2004.

```python
#5.1
run("""
SELECT
    SUM(amount)AS TotalPayment
FROM payments
WHERE 
    DAY(paymentDate)=28 AND
    MONTH(paymentDate)=10 AND
    YEAR(paymentDate)=2004
""")
```

```python
#5.2
run("SELECT * FROM payments WHERE YEAR(paymentDate)=2004 AND DAY(paymentDate)=28 AND MONTH(paymentDate)=10")
```

# 6. Report those payments greater than \$100,000.

```python
run("SELECT * FROM payments WHERE amount >100000")
```

# 9. What is the minimum payment received?

```python
run("SELECT customerNumber,checkNumber,paymentDate,MIN(amount)AS MinimumPayment FROM payments")
```

# 10. List all payments greater than twice the average payment.

```python
run("SELECT amount,AVG(amount)AS TwiceofAverage FROM payments GROUP BY paymentDate HAVING amount>TwiceofAverage")
```

```python
run("SELECT * FROM payments WHERE amount>(SELECT AVG(amount)*2 FROM payments)")
```

# 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT AVG((MSRP-buyPrice)/buyPrice*100) AS MarkupAvgPercentage FROM products")
```

# 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
run("SELECT CONCAT(firstName,' ',lastName) AS FullName,jobTitle FROM employees WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'")
```
