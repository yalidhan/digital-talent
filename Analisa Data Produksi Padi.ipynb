# 1. Deskripsi Data
Dataset ini berisi data produksi dari tanaman padi per provinsi yang dikumpulkan dari tahun ke tahun mulai dari tahun 1993 sampai dengan tahun 2013 oleh Badan Pusat Statistik. Data ini berisi 693 record data, dengan field/variabel data sebagai berikut :
1. tahun: Tahun
2. kode_provinsi: Kode Provinsi
3. nama_provinsi: Nama Provinsi
4. produksi: Total Produksi Tanaman Pangan (dalam Ton)

# 2.STATISTIK

```python
import pandas
data_padi =pandas.read_csv('padi.csv',delimiter=',',encoding="ISO-8859-1")
data_padi
```

```python
#Menampilkan Statistik produksi padi berdasarkan provinsi dan tahun
data_padi.groupby(["kode_provinsi","nama_provinsi","tahun"]).mean()
```

```python
#Jumlah produksi padi dari tahun 1993-2013 berdasarkan provinsi (dalam satuan Ton)
data_padi.groupby(["nama_provinsi"]).sum()['produksi']
```

```python
#Menampilkan Statistik produksi padi provinsi terendah berdasarkan tahun
data_padi.groupby(["tahun"]).min()["produksi"]
```

```python
#Menampilkan Statistik produksi padi provinsi tertinggi berdasarkan tahun
data_padi.groupby(["tahun"]).max()["produksi"]
```

```python
#Rerata produksi padi dari tahun 1993-2013 berdasarkan provinsi (dalam satuan Ton)
data_padi.groupby(["nama_provinsi"]).mean()['produksi']
```

```python
#Nilai tengah produksi padi dari tahun 1993-2013 berdasarkan provinsi (dalam satuan Ton)
data_padi.groupby(["nama_provinsi"]).median()['produksi']
```

```python
length=data_padi.produksi.count()
total=data_padi.produksi.sum()
miN=data_padi.produksi.min()
maX=data_padi.produksi.max()
mean=data_padi.produksi.mean()
median=data_padi.produksi.median()
modus=data_padi.produksi.mode()



print ("~~ STATISTIK KESELURUHAN~~")
print("Jumlah data :",length,"Data" )
print("Total produksi padi dari tahun 1993-2013 :",total,"Ton")
print("Produksi padi terendah :",miN,"Ton")
print("Produksi padi tertinggi:",maX,"Ton")
print("Rerata produksi padi:",mean,"Ton")
print("Nilai tengah produksi:",median,"Ton")
```

# 3. Visualisasi sederhana

```python
ax = data_padi.groupby(["tahun"]).mean()['produksi']
ax = ax.plot(kind='line', title ="Rata-Rata Produksi Padi Dari Tahun 1993 Sampai Dengan 2013",figsize=[20,20], legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

# 4. Narasi Analisis

Berdasarkan data di atas dapat disimpulkan bahwa sejak tahun 1993, produksi padi di Indonesia cenderung mengalami naik dan turun, namun sejak tahun 2006 produksi padi di Indonesia cenderung meningkat drastis hingga tahun 2013.
